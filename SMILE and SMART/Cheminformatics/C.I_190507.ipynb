### Introduction
* 1998년 F.K.브라운에 의해 정의가 되었다.
* Definition :  The use of informatics methods to solve chemical problem - by Johann Gasteiger(Father of chemoinformatics)
* 기술의 발전에 따라 만들 수 있고 분석할 수 있는 화합물의 수는 기하급수적으로 증가하는 반면 거대한 데이터를 효과적으로 활용할 수 있는 방법을 강구하기 위해 생성된 학문
> 기존의 다른 정보학과는 다르게 특수하게 화학은 분자를 데이터화 해야함<br>
> 분자는 원자와 결합, 또한 다른 분자에 따라 반응을 표현해야한다.
* 화학정보학(Chemoinformatics) 신약개발의 선도물질 확인 및 최적화 분야에서 더 빠르고 나은 결정을 내리기 위해서 데이터로부터 정보, 그리고 지식으로 전환해 나가기 위한 이들 정보자원의 혼합된 형태를 말한다.
* 주어진 화합물 Data로부터 Computer를 활용하여 구조와 성질간의 상관관계를 찾고, 이 관계를 통해 원하는 미지의 화합물의 성질을 예측, 미지의 성질을 가지는 화합물 설계
* Major Fields of Chemoinformatics
> * Structure Database
> * Quantitative Structure - Activity Relationship
> * Molecular Modeling
> * Structure Elucidation(분광화학적 예측) -> Chemometrics
> * Chemical Reaction and Synthesis Design<br>
> 5가지 범위에 걸쳐 화합물의 성질을 알아낸다.

# Chemoinformatics (Cheminformatics, Chemical Informatics) 2
### CT files(Chemical Table files)
> **CT files** ( Chemical Table files ) : 화합물에 필요한 구조정보들을 담고있는 파일. 가지고 있는 정보에 따라서 여러 형식이 있다.


* Molfiles : Ctab을 가지고 있는 기본 files ( .mol)
* rxnfiles : 반응을 표현하기 위해 사용하는 files
* SDfiles : 여러개의 Molfiles의 정보를 담고 있는 files ( .sd / .sdf )

### 가장 많이 쓰이는 files format(Molfiles, SDfiles)

* header : molecule의 이름이나, 저장된 시간을 나타내는 부분
* Ctab : 가장 기본적인 부분
* '`$$$$`' : 한 분자와 다른 분자를 나누는 경계를 표현




* Molfile의 마지막 부분은 M`  `END로 끝남

Block | Lines  | 
------|-------|
Header | Title | 
->| Program / file timestamp |  
->| Comment |  
Connection table | Counts line |  
->| Atom block |  
->| Bond block |  
->| Properties block |  
->| END line |  


1. 원자의 갯수
2. 결합의 갯수
3. 보통 사용하지 않는다.
4. 사용하지 않는다.
5. 구조이성질체의 중심(Chiral 중심)
6. `M END` 뒤에 특별히 명시되어 있다면 몇 줄에 걸쳐 있는지를 표현
7. properties block의 줄의 갯수
8. Molfile 형식의 version


* 3번(Z좌표)의 경우 특별히 계산된 값을 넣어주지 않는 한 python내에서는 0으로 표현된다.
* 7번의 경우 분자가 Chirality를 가진다면 R인지 S인지에 대한 명시해준다. ( 0 vs 1 )
* 8번의 경우 추가적으로 수소를 표현하지 않는 한 0으로 표현한다.
* 9번과 10번의 경우 특별한 경우에만 사용되므로 무시해도 좋다


* Bond Bolck에서 제대로 된 xyz의 좌표를 알고 있다면 5번을 제외한 나머지 line은 계산된다.
* 1,2,3번 line을 제외한 나머지 부분은 정확하게 계산을 하지 않는다면 무시해도 좋다.


* `M  END`를 포함한 줄의 갯수는 Countline의 7번에 갯수가 표현된다.
* $[1]$은 원자의 전하를 표현해준다
* $[2]$는 동위원소를 나타내준다
* 전하와 동위원소 이외에도 라디칼, 고리의 수 등을 넣을 수 있다.
