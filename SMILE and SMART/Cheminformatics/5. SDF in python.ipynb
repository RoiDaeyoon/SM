```python
# 폴더에 압축 파일 안에 sdf파일의 이름을 `pubchem.sdf`으로 바꿔서 입력하였다.
# sdf파일 中 compound의 갯수 counting

f = open('pubchem.sdf')
cnt = 0                  # 갯수를 세기 위해 임의로 지정
for line in f:           # f로 연 pubchem.sdf파일중 한개의 line을 지정
    if '$$$$' in line:
        cnt = cnt + 1
        
print(cnt)               #'$$$$'의 갯수가 몇개인지를 출력 즉 sdf가 몇개인지를 출력
```

```python
# sdf파일 속 25000개 中 1000개의 compound만 추출하여 사용
f = open('pubchem.sdf')
fo = open('pubchem1000.sdf','w')     #fo는 output 임의로 지정해도 상관없음
                                     # 'w'는 작성하겠다는 명령문
cnt = 0
sdf1=''
for line in f:
    sdf1 = sdf1 + line
    if '$$$$' in line :
        cnt = cnt + 1
        fo.write(sdf1)               # 한개의 compund가 종료될 시 sdf1에 작성
        sdf1=''
    if cnt ==1000:                   # '$$$$'가 1000개일 때 까지
        break
        
print(cnt)
```

```python
# 1000개의 compound의 ID만 추출
f = open('pubchem.sdf')
fo = open('pubchem1000.sdf','w')     #fo는 output 임의로 지정해도 상관없음
                                     # 'w'는 작성하겠다는 명령문
cnt = 0
sdf1=''
for line in f:
    sdf1 = sdf1 + line
    if '$$$$' in line :
        cnt = cnt + 1
        fo.write(sdf1)              # 한개의 compund가 종료될 시 sdf1에 작성
        print(sdf1.split('\n')[0])  #분자의 ID에 해당하는 번호만 출력(첫 번째 줄)
        sdf1=''
    if cnt ==1000:
        break
        
print(cnt)
```

```python
# 분자의 Chemical table까지 출력
f = open('pubchem.sdf')
fo = open('pubchem1000.sdf','w')     #fo는 output 임의로 지정해도 상관없음
                                     # 'w'는 작성하겠다는 명령문
cnt = 0
sdf1=''
for line in f:
    sdf1 = sdf1 + line
    if '$$$$' in line :
        cnt = cnt + 1
        sdf2 = sdf1.split('M  END')[0]        #Molfile이 끝나는 지점에서 나눈다.
        sdf2 = sdf2 + 'M  END\n' + '$$$$\n'   # \n enter키(띄어쓰기)
        fo.write(sdf2)                        # 한개의 compund가 종료될 시 sdf2에 작성
        sdf1=''
        print(sdf2)
    if cnt ==1000:
        break
        
print(cnt)
```

```python
# ID와 원자의 갯수 세기
f = open('pubchem.sdf')
fo = open('pubchem1000.sdf','w')       #fo는 output 임의로 지정해도 상관없음
                                       # 'w'는 작성하겠다는 명령문
def atomnumber(sdf1):
    atomid = sdf1.split('\n')[0]      # sdf1의 첫 번째 띄어쓰기
    atomline = sdf1.split('\n')[3]    # molfile의 CT 중 첫 번째 행
    atom = atomline.split()[0]        # 첫 번째 행의 첫 번째 원소는 원소의 갯수
    print(atomid, atom)
    
    
cnt = 0
sdf1=''
for line in f:
    sdf1 = sdf1 + line
    if '$$$$' in line :
        cnt = cnt + 1
        sdf2 = sdf1.split('M  END')[0]       #Molfile이 끝나는 지점에서 나눈다.
        sdf2 = sdf2 + 'M  END\n' + '$$$$\n'
        fo.write(sdf2)                       # 한개의 compund가 종료될 시 sdf1에 작성
        atomnumber(sdf1)
        sdf1=''
    if cnt ==1000:
        break
        
print('total= ', cnt)
```

```python
# atomnumber 함수의 원자의 숫자를 atomnum에 숫자의 string으로 작성
f = open('pubchem.sdf')
fo = open('pubchem1000.sdf','w')     #fo는 output 임의로 지정해도 상관없음
                                     # 'w'는 작성하겠다는 명령문
def atomnumber(sdf1):
    atomid = sdf1.split('\n')[0]     # sdf1의 첫 번째 띄어쓰기
    atomline = sdf1.split('\n')[3]   # molfile의 CT 중 첫 번째 행
    atom = atomline.split()[0]       # 첫 번째 행의 첫 번째 원소는 원소의 갯수
    print(atomid, atom)
    return int(atom)                 # 출력시 atom값을 숫자로 출력
    
cnt = 0
sdf1=''
atomnum=[]
for line in f:
    sdf1 = sdf1 + line
    if '$$$$' in line :
        cnt = cnt + 1
        sdf2 = sdf1.split('M  END')[0]        # Molfile이 끝나는 지점에서 나눈다.
        sdf2 = sdf2 + 'M  END\n' + '$$$$\n'
        fo.write(sdf2)                        # 한개의 compund가 종료될 시 sdf1에 작성
        atomnum.append(atomnumber(sdf1))
        sdf1=''
    if cnt ==1000:
        break

print(atomnum)        
print('total= ', cnt)
```

```python
# 첫 번째부터 1000 번째 까지의 molecule의 원자의 갯수를 가시화
import matplotlib.pyplot as plt
plt.xlabel('Molecules')
plt.ylabel('Number of Atoms')
plt.title('Chemoinformatics')
x = range(len(atomnum))
plt.plot(x,atomnum)
plt.show()
plt.savefig('Myplot.png')
```
