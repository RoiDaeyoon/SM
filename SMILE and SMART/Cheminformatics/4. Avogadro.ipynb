# Avogadro

- SAMSON 같이 화합물의 3D 구조를 관찰하고, 화합물을 간단히 만들고 저장할 수 있는 프로그램.


* 첫 화면 - 여기서 빨간 네모 박스 안의 툴을 이용해 간편하게 화합물을 만들거나, 원하는 각도로 돌리거나 최적화, 길이 및 각도의 측정을 할 수있다.
* 또는 `Mol file, SD file` 의 파일을 열어서 불러올 수 있다.




## Tool 박스



왼쪽부터 순서대로 

1. Draw Tool - 화합물을 만들 수 있다.
2. Navigation Tool - 화합물을 3차원으로 관찰하고, 이동시킬 수 있다.
3. Bond Centric Manipulation Tool - 특정 결합을 기준으로 나머지 구조를 회전시킬 수 있다.
4. Manipulation Tool - 
5. Selection Tool - 화합물을 선택할 수 있다.
6. Auto Rotation Tool - 자동회전시킬 수 있다.
7. Auto Optimization Tool - 임의로 만들 화합물의 결합각, 길이를 에너지가 가장 낮은 상태를 계산해서 변형시켜준다.
8. Click to Measure - 원자간의 길이, 3원자들이 이루는 각을 측정할 수 있다.
9. Align Molecules - 특정 원자들의 축을 원하는 축 ( x, y, z ) 에 평행하게 바꿔준다.

### 1. Draw Tool

드래그 앤 드랍으로 $ C_{2}H_{6}$ 을 간단하게 만들 수 있다.

좌측의 `Draw Settings` 에서 원자, 결합의 종류, 수소를 갖는지 여부를 설정 할 수 있다. <br>




또는 마우스의 좌클릭으로 만들고, 우클릭으로 클릭된 원자를 지울 수 있다.<br>





### 2. Navigation Tool

Tool 박스의 `Navigation Tool` 을 이용해서 마우스의 좌클릭 (회전) , 우클릭 (이동)으로 바꿀 수 있다.




### 3. Bond Centric Manipulation Tool

> 원자를 클릭하면 그 원자가 주위의 원자들과 이루는 각도를 보여준다.



<br>
> 결합을 클릭하면 그 결합을 포함하는 평면을 보여준다.



> 이 상태에서 주위의 원자들을 클릭하면 또 다른 평면을 보여주고, 클릭된 결합을 축으로 회전할 수 있다.



<br>

> 또는 결합을 이루는 원자들을 클릭해 드래그 하면 결합한 다른 원자를 기준으로 회전한다.





### 4. Manipulation Tool
특정 원자를 끌어다 옮기거나, 좌측에서 x,y,z축 이동, 회전을 할 수 있다.





### 5. Selection Tool
원자, 결합 또는 부분을 선택하거나 화합물의 중심( 원자, 무게 )을 볼 수 있다.





### 6. Auto Rotation Tool

마우스로 드래그 & 드랍으로 자동 회전하도록 하거나, xyz축에 설정으로 회전시킬 수 있다.



### 7. Auto Optimization Tool

임의로 만든 화합물의 에너지를 계산하여 최적화된 결합의 길이, 각도를 계산하여 바꾼다. <br>
좌측에서 *Froce Field*, 스텝, 등을 설정할 수 있다. <Br>
메뉴 탭의 `Extensions > Optimize Geometry`와 같은 기능이다.







### 8. Click to Measure

클릭된 원자들간의 결합길이, 각도, `Dihedral`을 측정한다.







### 9. Align Molecules
선택된 두개의 원자를 하나의 축으로 생각하여, 이를 x,y,z축에 평행하게 회전시킨다.








## Display Settings
다른 방식의 분자들로도 관찰할 수 있다.







### `Build > Cartesian Editor`에서 각 원자들의 x,y,z 좌표를 확인, 수정하여 더 정밀한 화합물을 만들 수 있다.



### `Build > Insert > ...` 복잡한 구조의 화합물들을 쉽게 만들 수 있다.

> SMILES 입력으로 화합물 만들기

<br>




> `> Fragment` 로 널리 알려진 구조식들을 불러 올 수 있다.





### `Extensions > Optimize Geometry(Ctrl+Alt+O)` 분자들간의 기하학적 구조를 안정화해준다

>예시) 임의로 생성한 헥산



>Optimize Geometry후의 헥산


### `Extensions > Molecular Mechanics > Conformer Search` 가장 안정한 기하학적 구조로 설정해준다
>다만 각 원자가 지닐 수 있는 회전의 경우의 수가 많다면 가동 시간이 오래걸린다.<br>




>Conformer Search를 끝낸 가장 안정한 구조


### `View > Properties > Molecule Properties..` 분자의 물성을 표현해준다.
>다만 Avogadro 내부에 저장된 분자에 관해서만 출력이 된다.<br>
