### 1. Read / Write


```python
#RDkit 적용
from __future__ import print_function #2.X대 python호환을 위해
from rdkit import Chem
from rdkit.Chem import AllChem
ala = 'CC(C(=O)O)N' #Alanine의 단순한 string
ala1 = 'O=C(O)[C@@H](N)C' #Alanine의 Chirality 포함
m = Chem.MolFromSmiles(ala)
m1 = Chem.MolFromSmiles(ala1)
```

```python
print(m) # Mol object가 출력이됨(mole의 주소)
print(m1)
```

```python
# 제대로 mole이 저장되어 있는지 확인하기
# MolBlock : Chemical Table
print(Chem.MolToMolBlock(m))
```

```python
print(Chem.MolToMolBlock(m1))
```

```python
print(Chem.MolToSmiles(m))
print(Chem.MolToSmiles(m1))
```

```python
print(Chem.MolToSmarts(m))
print(Chem.MolToSmarts(m1))
```

```python
print(Chem.MolToInchi(m))
print(Chem.MolToInchi(m1))
```

```python
# Mol fils 형식으로 저장
f = open('ala.mol','w')
f.write(Chem.MolToMolBlock(m))
f.close()
```

```python
f = open('ala1.mol','w')
f.write(Chem.MolToMolBlock(m1))
f.close()
```

### 5. Despcriptor Calculations(물성 계산)

```python
from rdkit.Chem import Descriptors
# 분자량 계산
print(Descriptors.MolWt(mh))
```

```python
#wildman-crippen method를 이용한 LogP 계산
print(Descriptors.MolLogP(mh))
```

$ P = \frac{옥탄올 층의 화합물의 농도}{물 층의 화합물의 농도}$
* LogP는 섞이지 않는 두 용매인 물과 옥탄올(octanol)에 화합물을 녹였을 때 물과 옥탄올층에 녹아 있는 화합물 농도의 비를 분배계수로 나타낸 것
* 화합물이 친수성(hydrophilicity)을 갖는지 소수성(hydrophobicity)을 갖는지 예상할 수 있으며 값이 클수록 소수성이 큰 것으로 예상
* 일반적으로 임상실험을 통과한 약들의 경우, 90%가량이 0 - 5 사이의 logP값을 나타낸다

```python
print(Descriptors.RingCount(mh))
```

```python
print(Descriptors.TPSA(mh))
```

**TPSA; Topological Polar Surface Area(위상극면적)**

* Topological polar surface area (TPSA) of a molecule is defined as the surface sum over all polar atoms or molecules, primarily oxygen and nitrogen, also including their attached hydrogen atoms.
* TPSA는 모든 극성 원자 또는 분자(주로 산소 혹은 질소) 그리고 그것들에 부착된 수소원자를 포함하는 표면적의 합계로 정의된다.
* Commonly used medicinal chemistry metric for the optimization of a drug's ability to permeate cells.
* 주로 약용 화학에서 약들의 투과능력을 최적화하기 위해 사용된다.
* Molecules with a polar surface area of greater than 140 angstroms squared tend to be poor at permeating cell membranes.
* 140 옹스트롬의 제곱면적보다 큰 극면적을 가지는 분자는 세포막 투과능력이 떨어지는 경향이 있다.
