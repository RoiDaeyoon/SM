### 2D Rotation in Cartesian

기존의 x,y 축에서 $ \theta $ 만큼 회전한 새로운 x', y' 축에서의 좌표를 기존의 좌표 (x, y)와 삼각함수를 이용해 구한다. 



$\binom{X'}{Y'} = \begin{pmatrix}
cos\theta  &sin\theta  \\ 
-sin\theta  & cos\theta 
\end{pmatrix}\binom{X}{Y}$

### 3) Cartesian -> Internal

> A를 원점으로 하는 임의의 직교좌표계에서 벡터의 성질을 이용해 Geometry 정보를 얻을 수 있다.






> 1. $ \vec{BA} \cdot \vec{BC} = \cos{\alpha}\left | \vec{BA} \right | \left | \vec{BC} \right | $
2. $ \vec{BA} \times \vec{BC} = \vec{N} $  (평면 ABC에 수직인 벡터)
3. $ \vec{CD} \times \vec{CB} = \vec{Q} $  (평면 BCD에 수직인 벡터)
4. $ \vec{N} \cdot \vec{Q} = \cos{\omega}\left | \vec{N} \right | \left | \vec{Q} \right |  $ 

### 4) Internal -> Cartesian

> 주어진 결합길이, 각도, $\omega$를 이용해 직교좌표계로 변환한다.




> D의 좌표를 C를 기준으로 하는 다른 좌표계로의 Translation을 이용해 구할 수 있다.
