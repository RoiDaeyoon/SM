### Introduction
* 1998년 F.K.브라운에 의해 정의가 되었다.
* Definition :  The use of informatics methods to solve chemical problem - by Johann Gasteiger(Father of chemoinformatics)
* 기술의 발전에 따라 만들 수 있고 분석할 수 있는 화합물의 수는 기하급수적으로 증가하는 반면 거대한 데이터를 효과적으로 활용할 수 있는 방법을 강구하기 위해 생성된 학문
> 기존의 다른 정보학과는 다르게 특수하게 화학은 분자를 데이터화 해야함<br>
> 분자는 원자와 결합, 또한 다른 분자에 따라 반응을 표현해야한다.
* 화학정보학(Chemoinformatics) 신약개발의 선도물질 확인 및 최적화 분야에서 더 빠르고 나은 결정을 내리기 위해서 데이터로부터 정보, 그리고 지식으로 전환해 나가기 위한 이들 정보자원의 혼합된 형태를 말한다.
* 주어진 화합물 Data로부터 Computer를 활용하여 구조와 성질간의 상관관계를 찾고, 이 관계를 통해 원하는 미지의 화합물의 성질을 예측, 미지의 성질을 가지는 화합물 설계
* Major Fields of Chemoinformatics
> * Structure Database
> * Quantitative Structure - Activity Relationship
> * Molecular Modeling
> * Structure Elucidation(분광화학적 예측) -> Chemometrics
> * Chemical Reaction and Synthesis Design<br>
> 5가지 범위에 걸쳐 화합물의 성질을 알아낸다.

* 화학정보학에서 가장 중요한 부분은 분자 구조를 어떻게 표현을 하고 입력을 하는 것이다.
* 목적
> 현재 알려져 있지 않지만 성능이 좋은 약물이 존재한다. 그러한 성질을 가진 약물을 합성하기 위해 chemoinformatics를 이용한다.


```python
from __future__ import print_function #2.X대 python호환을 위해
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
```

```python
# 알라닌(Alanine) :  HO2CCH(NH2)CH3
m = Chem.MolFromSmiles('[C@H](N)C(O)=O')
```

```python
#Chemical Table 추후 설명
Chem.MolToMolBlock(m)
```

* 분자의 성질 및 구조를 컴퓨터가 처리할 수 있는 언어(1차원 string)로 사용하기 위해 여러가지 시도(WLN, ROSDAL,etc.)가 있으나 그 중 SMILES가 Chemoinformatics분야에서 최초의 언어이고 통상적으로 사용하는 언어이다.
* SMILES의 특징
> 한 가지의 화합물을 여러가지 방법으로 표현할 수 있다.<br>
> 코드의 string을 보면 직관적으로 어떤 화합물인지 유추가 비교적 쉽다.<br>
> 자세한 내용은 SMILES folder내 jupyter notebook 참조
* InChI(International Chemical Identifier)의 특징
> 한 가지의 화합물은 단 한 가지의 방법으로만 표현 가능하다.<br>
> 다만 복잡한 형태의 코드이기 때문에 코드를 보더라도 직관적으로 유추하기가 힘들다.<br>
